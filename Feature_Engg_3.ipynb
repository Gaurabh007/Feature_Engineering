# SMOTE(Synthetic Minority Oversampling Technique)

SMOTE is a technique used in Machine Learning to handle the imbalanced Dataset where the minority classes have significantly lesser instance than majority classes. It involves generating synthentic data points of the minority class by interpolating (creating our own synthetic data) between the existing instances.

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=10)          #creating the imbalanced dataset
```

```python
x.shape,y.shape
```

```python
import pandas as pd
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
final_df
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
x,y = oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```
